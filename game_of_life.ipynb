```python
%pylab inline
from FSMSIM import *
from IPython.display import display, Markdown, clear_output
import ipywidgets as widgets
```

```python
loader = ModuleLoader()
loader.load("game_of_life.fsm")
```

```python
gol = loader.modules["GameOfLife"].init()
SIZE = 17
init = np.random.randint(0, 2, SIZE * SIZE).astype(str)
gol.inputs["i"].set("".join(init))
gol.tick()
```

```python
def visualize_game(gol):
    board = np.array(list(gol.outputs["o"].evaluate())).astype(float).reshape((SIZE, SIZE))
    plt.imshow(board, cmap="gray")
    plt.show()
```

```python
while True:
    clear_output()
    visualize_game(gol)
    gol.tick()
```
